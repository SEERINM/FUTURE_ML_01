```python
import pandas as pd
df = pd.read_csv("C:\\Users\\SEERIN MUBEENA CSE\\Desktop\\Future Interns\\train.csv")
print(df.head())
```

```python
print((df.columns))
print(df.info)
```

```python
print(df.isnull().sum())
```

```python
# Convert 'Order Date' to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
# Group sales by month-year
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum().reset_index()
monthly_sales['Order Date'] = monthly_sales['Order Date'].dt.to_timestamp()
# Rename columns to match Prophet's requirements
monthly_sales.rename(columns={'Order Date': 'ds', 'Sales': 'y'}, inplace=True)

# Preview the result
print(monthly_sales.head())
```

```python
'''This prepares your data for Prophet by making sure:

ds is your date

y is your sales'''
```

```python
from prophet import Prophet
import matplotlib.pyplot as plt

# Initialize the Prophet model
model = Prophet()

# Fit the model with the monthly sales data
model.fit(monthly_sales)

# Generate future dates (6 months ahead)
future = model.make_future_dataframe(periods=6, freq='M')

# Forecast future sales
forecast = model.predict(future)

# Plot the forecast
model.plot(forecast)
plt.title("Superstore Sales Forecast")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
# Keep only relevant columns for dashboarding
export_df = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]

# Merge with actuals to compare real vs predicted
export_df = export_df.merge(monthly_sales, on='ds', how='left')

# Rename columns for clarity in Power BI
export_df.rename(columns={
    'ds': 'Date',
    'yhat': 'Forecasted Sales',
    'yhat_lower': 'Lower Estimate',
    'yhat_upper': 'Upper Estimate',
    'y': 'Actual Sales'
}, inplace=True)

# Save to CSV
export_df.to_csv("superstore_sales_forecast.csv", index=False)

print("✅ Export complete. File ready for Power BI: superstore_sales_forecast.csv")
```
